### Matrix Multipication

```python
def multiplyMatrix(X, Y):
    
    rowsX,colX = X.shape
    rowsY,colY = Y.shape
    
    # checking the dimensions of matrices
    if(colX!=rowsY):
        return 'Dimensions not matched'
    
    else:
        output =np.zeros((rowsX,colY))
        
        for i in range(rowsX): # for selecting rows of first matrix
            
            for j in range(colY): # for selecting columns of second matrix
                
                for k in range(colX): # for multiplying row to column
                    
                    output[i][j] += X[i][k]*X[j][k]
                    
    return output
    
```

```python
A = np.random.normal(2,0.01,size=(100,20)) # random vector with normal distribution
v = np.random.normal(2,0.01,size=(20,1)) # random vector with normal distribution
c = multiplyMatrix(A,v)
print(' Output vector c:',c)
print(' Dimensions of vector c:',c.shape)
```

```python
# function for calculation in mean and standard deviation
def SdMeanCal(c):
    m=0
    s=0
    
    for i in range(len(c)): # summing the values of vector
        m += c[i]
    mean = m/len(c) # divinding the sum with total size of vector
    
    for i in range(len(c)):
        s += (c[i]-mean)**2 # subracting mean from each value and then squaring and sum 
    sd = (s/len(c))**0.5 # dividing with total size and then taking square root
    
    return np.squeeze(mean),np.squeeze(sd) # changing the shape to scalar.
    
```

```python
mean,sd = SdMeanCal(c)
print('Mean :' , mean)
print('Stamdard Deviation :' , sd)
```

```python
# plotting the histogram with 5 bins with mean value in centre
n_bins = 5
fig, axs = plt.subplots(1,1)
axs.hist(c, bins=n_bins)
plt.title('histogram of vector c')
```

```python
# defining a diction for grading systems
grades = {
                (0, 55) : 'F',
                (56, 59) : 'D',
                (60, 65) : 'C',
                (66, 69) : 'C+',
                (70, 75) : 'B-',
                (76, 79) : 'B',
                (80, 85) : 'B+',
                (86, 89) : 'A-',
                (90, 95) : 'A',
                (96, 100) : 'A+',
                }

# defining a function for getting a grade from dictionary for a given point
def getGrade(point):
    result = 'None'
    
    # iterating through grades dictionary
    for (k1, k2) in grades:
        
        # checking if point lies between the dictionary key
        if (k1 <= int(point+0.5) and k2 >= int(point+0.5)): # adding 0.5 for rounding off values above .5
            result =  grades[(k1,k2)] # assigning grade if value is in between
            break
            
    return result
     
# calculating grade column from Average Point Column and for each row of AveragePoint columnn mapping getGrade() with lamda funtion.      
df['grades']= df['AveragePoint'].map(lambda x: getGrade(x))

df.head(10)
```

## Exercise 2: Linear Regression through exact form.

#### • Generate 3 sets of simple data; a matrix A with dimensions 100 × 2. Initialize it with normal distribution μ = 2 and σ = [0.01, 0.1, 1]

```python
np.random.seed(0)
rows = 100
col = 2
mean, sd = 2, 0.01
A = np.random.normal(mean, sd, (rows,col))
```

#### • Implement LEARN-SIMPLE-LINREG algorithm and train it using matrix A to learn values of β0 and β1

```python
def LEARN_SIMPLE_LINREG(A,n):
    
    numerator=0
    denominator=0
    X= A[:, 0] # first column as X
    Y= A[:, 1] # second column as target Y
    Xmean= np.mean(A[:, 0])
    Ymean= np.mean(A[:, 1])
    
    for i in range (n): # for calculating B1
        numerator += (X[i] - Xmean) * (Y[i] - Ymean)
        denominator += (X[i] - Xmean) ** 2
        
    B1 =numerator/denominator
    B0= Ymean - (B1 * Xmean)
    
    return B0,B1,X,Y
```

#### • Implement PREDICT-SIMPLE-LINREG and calculate the points for each training example in matrix A.

```python
def predict_simple_linreg(X,B0,B1):
    
    y_hat= B0 + (B1 * X) # output of simple linear model
    
    return y_hat
```

```python
B0,B1,X,Y=LEARN_SIMPLE_LINREG(A,rows) # learning values of B0,B1
Y_hat = predict_simple_linreg(X,B0,B1) # predicting output for the model
print("b0 =", B0)
print("b1 =", B1)
```

#### • Plot the training data (use plt.scatter) and your predicted line (use plt.plot).

```python
plt.plot(X,Y_hat,color ='#6c1abd') # regression line for targets
plt.scatter(X,Y) # for training points
```

#### • Put β0 to zero and rerun the program to generate the predicted line. Comment on the change you see for the varying values of σ

```python
# For Standard Deviation = 0.01
np.random.seed(0)
mean, sd = 2, 0.01
A = np.random.normal(mean, sd, (rows,col))
B0,B1,X,Y=LEARN_SIMPLE_LINREG(A,rows)
Y_hat = predict_simple_linreg(X,0,B1)
plt.subplot(1, 3, 1)
plt.plot(X,Y_hat,color ='#6c1abd')
plt.scatter(X,Y)
plt.title("SD = 0.01")

# For Standard Deviation = 0.1
mean, sd = 2, 0.1
A = np.random.normal(mean, sd, (rows,col))
B0,B1,X,Y=LEARN_SIMPLE_LINREG(A,rows)
Y_hat = predict_simple_linreg(X,0,B1)
plt.subplot(1, 3, 2)
plt.plot(X,Y_hat,color ='#6c1abd')
plt.scatter(X,Y)
plt.title("SD = 0.1")

# For Standard Deviation = 1
mean, sd = 2, 1
A = np.random.normal(mean, sd, (rows,col))
B0,B1,X,Y=LEARN_SIMPLE_LINREG(A,rows)
Y_hat = predict_simple_linreg(X,0,B1)
plt.subplot(1, 3, 3)
plt.plot(X,Y_hat,color ='#6c1abd')
plt.scatter(X,Y)
plt.title("SD = 0.1")
```

<b>Setting B0 = 0 , the regrressor line goes through 0 on y axis as B0 is the intercept. Also with variation of SD, points becomes more scatter as increasing SD and slope of regressor line changes to best fit the points.</b>

#### • Use numpy.linalg lstsq to replace step 2 for learning values of β0 and β1. Explain the difference between your values and the values from numpy.linalg lstsq.(1 point)

```python
# initializing the Data with normal distribution
np.random.seed(0)
rows = 100
col = 2
mean, sd = 2, 0.01
Data = np.random.normal(mean, sd,(rows,col))

x=Data[:,0].reshape((100,1)) # first column as input
y=Data[:,1] # second column as target#

bias_column = np.ones(shape=(100,1)) # bias column of ones
A = np.append(bias_column,x,axis=1)  # adding bias column with input for x0 feature

beta0,beta1 = np.linalg.lstsq(A, y,rcond=-1)[0] # linear regressor from numpy
print("b0 =", beta0)
print("b1 =", beta1)

Y_hat =(beta0 + beta1 * x) # predicting output

# plotting 
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,Y_hat)
plt.scatter(x,y)
plt.show()
```
